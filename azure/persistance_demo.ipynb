```python
df=spark.read.format("csv").option("header","true").option("inferSchema","true").load("/Volumes/avd_azure_db/default/raw_data/emp.csv")
df.show()
```

```python
from pyspark.sql.functions import col 
df2=df.withColumn("tax", col("salary") * 0.05)

df2.show()
```

```python
df3=df2.filter(col("address") == "Mumbai")
df3.show()
```
